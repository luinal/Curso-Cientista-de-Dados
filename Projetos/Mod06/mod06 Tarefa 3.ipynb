```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("Resources/sinasc_RO_2019.csv")
df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
quebras = [
    (df['APGAR5'] >= 8),
    (df['APGAR5'] >= 6) & (df['APGAR5'] < 8),
    (df['APGAR5'] >= 4) & (df['APGAR5'] < 6),
    (df['APGAR5'] < 4)
]

categorias = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']

df['Classificação'] = np.select(quebras, categorias, default='Unknown')
df.head()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
df.columns
```

```python
g = sns.catplot(x='APGAR5', y='GESTACAO', 
                hue='SEXO', 
                data=df, 
                kind='box', 
                height=5, 
                aspect=1.5, 
                legend_out=True
)

g.set_axis_labels("APGAR5 Score", "Gestação")
g.add_legend(title="Sexo")
plt.title('Distribution of APGAR5 by Sex Across Different Gestation Periods')
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
crosst = pd.crosstab(df['GESTACAO'], df['APGAR5'])
crosst
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
crosst.sum(axis = 1)

crosst_pctlin = crosst.div(crosst.sum(axis = 1), axis = 0)
crosst_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
crosst_pctlin.plot.bar()
```

O gráfico mostra APGAR5 alto a partir das 32 semanas, e baixo antes disso, portanto corrobora parcialmente com a afirmação, ou pode haver alguma outra variável influenciando o APGAR5 durantes as 32~36 semanas.
